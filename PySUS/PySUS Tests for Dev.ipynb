### Listando Informaçoes Agregadas - IBGE

```python
import pandas as pd
from pysus.online_data import IBGE

import ipywidgets as widgets
%matplotlib inline
```

```python
## List Available IBGE Surveys Informations

ag = IBGE.list_agregados()
ag
```

```python
## Listing Data from Demographic Survey

pd.DataFrame(ag[ag.id=='CM'].agregados.iloc[0])
```

```python
## Listing Population Projections

pd.DataFrame(ag[ag.id=='XE'].agregados.iloc[0])
```

```python
## Downloading Data 

opts= [(r.nome, int(r.id)) for r in pd.DataFrame(ag[ag.id=='CM'].agregados.iloc[0]).itertuples()]
ds = widgets.Dropdown(
    options=opts,
    value=475,
    description='Number:',
)
display(ds)
IBGE.metadados(ds.value)
```

```python
IBGE.lista_periodos(475)
```

```python
IBGE.localidades_por_agregado(475,nivel='N3')
```

```python
ds = IBGE.FetchData(475,periodos=1996,variavel=93,localidades='N3[all]',
                              classificacao='58[all]|2[4,5]|1[all]',view='flat')
ds.JSON[0]
```

```python
ds.to_dataframe()
```

### Using SIDRA Endpoint

```python
# IBGE also has a simpler API at https://api.sidra.ibge.gov.br that PySUS also gives access through 
# a simple function. Below we have table 200, which is a sample from the resident population. 
# classification 2 is sex, of which I am fetching all categories: total, Homens, and Mulheres. 
# Terrotorial level 6 is municipality.


df = IBGE.get_sidra_table(200,territorial_level=6,classification=2, categories='all')
df
```

```python

# Suppose we just wanted a single municipality, Balneário Rincão (SC): 
# Unfortunately for this one there is no data available, thus the ... in the column Valor. 
# Try another one.

single = IBGE.get_sidra_table(200,territorial_level=6, geocode=4220000,period='last',classification=2, categories='all')
single
```
