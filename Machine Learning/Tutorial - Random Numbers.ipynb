```python
## Dice Simulation - it produces 100 randomic numbers (list) from 1 to 6. 
## Remember that Python is inclusive, so the upper limit must be +1 (higher)

die_rolls1 = np.random.randint(1,7,5000)
print(die_rolls1)
```

```python
## Same as example abobve, but generating floating Numbers

die_unif = np.random.uniform(1,7,5000)
print(die_unif)
```

```python
# DICE Simulation but using normal distribution (Bell Curve) - it will summarize the die_rolls1 + the new generated dir_rools_2

die_rolls2 = np.random.randint(1,7,5000)
combo_rolls = zip(die_rolls1,die_rolls2)
combos = [a + b for a, b in combo_rolls]
print(combos)

```

```python
## Plot the Curve - we will see the Normal Distribution happening 

plt.hist(combos,100)
plt.show()
```

```python
## Generate Randoms respecting Bell's Normal Distribution 

combo_rolls = np.random.normal(7,1,100)
combo_rolls
```

```python
plt.hist(combo_rolls,100)
plt.show()
```

```python
## Sample for a Normal Distribution with mean (mu) and standard deviation (sigma)

def corr2cov(p: np.ndarray, s: np.ndarray) -> np.ndarray:
    """Covariance matrix from correlation & standard deviation)"""
    d=np.diag(s)
    return d @ p @ d
```

```python
corr = np.array([[1,0.9],[0.9,1]])
stdev = np.array([14,25])
mean=np.array([38,10])
cov=corr2cov(corr,stdev)
print(cov)
data = np.random.multivariate_normal(mean=mean,cov=cov,size=500)
data[:10]

```

```python
X,Y = zip(*data)
plt.scatter(X,Y)
```
